```python
import os
#from pprint import pprint
import dask.dataframe as dd
import matplotlib.pyplot as plt
import numpy as np
import aux as aux
import anomalies as an
```

```python

path=os.path.abspath('./data/data.csv')
df = dd.read_csv(path)
```

```python
hardware_list = ['SW-065', 'SW-088', 'SW-106', 'SW-115']

engine_list = ['S117', 'S118', 'S169', 'S170']
```

```python
hwid = 'SW-106'
for eng in engine_list:
    print("Hardware :" , hwid, "; Engine :", eng)
    time, anom, val = an.anomaly_column(df, hwid, eng, aggfunc='max')
    an.plot_anom_or_trend(time, anom, val, 'both')
```

```python
hwid = 'SW-106'
binary_metrics = ['S112','S113','S114','S115','S123','S127', 'S130','S171','S172','S173','S174','S179']

for bin in binary_metrics:
    print("Hardware :" , hwid, "; Engine :", bin)
    time, anom, val = an.anomaly_column(df, hwid, bin, aggfunc='max')
    an.plot_anom_or_trend(time, anom, val, 'both')
```
