```python
import os
#from pprint import pprint
import dask.dataframe as dd
import matplotlib.pyplot as plt
from dask.distributed import Client
import numpy as np
import aux as aux
import anomalies as an
import pandas as pd
```

```python
client = Client('172.18.0.2:8786')
```

```python
path=os.path.abspath('./data/data.csv')
df = dd.read_csv(path)
```

```python
hardware_list = ['SW-065', 'SW-088', 'SW-106', 'SW-115']

engine_list = ['S117', 'S118', 'S169', 'S170']
```

```python
for hwid in hardware_list:
    for eng in engine_list:
        print("Hardware :" , hwid, "; Engine :", eng)
        time, anom, val = an.anomaly_column(df, hwid, eng, aggfunc='max')
        an.plot_anom_or_trend(time, anom, val, 'anom', hwid, eng)
    
        results_df = pd.DataFrame({
            'Time': time,
            'Anomaly': anom,
            'Value': val
        })
    
        filename = f'{hwid}_{eng}.csv'
        filepath = os.path.join('Motors_df', filename)
        
        results_df.to_csv(filepath, index=False)
```

```python
client.close()
```
